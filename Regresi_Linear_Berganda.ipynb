````python
from google.colab import files
import pandas as pd

# Langkah 1: Upload File
uploaded = files.upload()

# Langkah 2: Membaca file yang diupload (pastikan nama file sesuai)
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

# Tampilkan 5 baris pertama untuk verifikasi
print("Data berhasil dimuat:")
display(df.head())
````

````python
import statsmodels.api as sm

# Menentukan Variabel Independen (X) dan Dependen (Y)
X = df[['Inflasi (%)', 'Kurs USD/IDR']]
y = df['BI_Rate (%)']

# Menambahkan konstanta (intercept) agar model lebih akurat
X = sm.add_constant(X)

# Membuat model OLS (Ordinary Least Squares)
model = sm.OLS(y, X).fit()

# Menampilkan Ringkasan Hasil Regresi
print("--- HASIL REGRESI LINEAR BERGANDA ---")
print(model.summary())
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih kolom numerik saja untuk korelasi
correlation_matrix = df[['BI_Rate (%)', 'Inflasi (%)', 'Kurs USD/IDR']].corr()

# Membuat Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Variabel Makro')
plt.show()
````
